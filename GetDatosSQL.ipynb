```python
import os
import pyodbc as pyodbc
import pandas as pd
import tkinter
from tkinter import * 
from tkinter.ttk import *
from tkinter import messagebox
from tkinter import BOTH, END, LEFT


class Adder(ttk.Frame):
    """The adders gui and functions."""
    def __init__(self, parent, *args, **kwargs):
        ttk.Frame.__init__(self, parent, *args, **kwargs)           
        self.root = parent
        self.init_gui()

    def calculate(self):
        global conn        
        firstname = str(self.first_entry.get())
        lastname = str(self.last_entry.get())   
        
        

        if (firstname and lastname):  # "You entered first name and last name"
            
            try:
                conn = pyodbc.connect('Driver={SQL Server};Server=localhost;Database=Safiro;Trusted_Connection=yes;')
                
            except pyodbc.Error as ex:
                sqlstate = ex.args[0]
                if sqlstate == '28000':
                    self.answer_label['text'] = "No tienes acceso a la BD." 
            cursor = conn.cursor() 
            SQLCommand = ("SELECT * "      
                "FROM dbo.Ingresos_salidas_telework_user "   # table name
                "with (nolock)"
                "WHERE CONVERT(DATE,fecha_hora_entrada) >= ? AND CONVERT(DATE,fecha_hora_entrada) <= ?")          
            
            
            cursor.execute(SQLCommand,firstname,lastname)           
            
            
            columns = [i[0] for i in cursor.description]  ##obtenemos los encabezados de las columnas           
            
            data = cursor.fetchall()           
            print(columns)
            print(data)
            #df = pd.DataFrame(list(data), columns=columns)    
            #print(df)
            
            
            
            #writer = pd.ExcelWriter('C:\Users\LENOVO\Desktop\SafiroMarcaciones.xlsx', sheet_name = "SafiroMarcaciones", header = True)
            #df.to_excel(writer, sheet_name='Marcaciones')
            #writer.save()
            
           
            
            results = cursor.fetchmany(10)
            if results:
                self.output0.delete(0, END)
                self.output0.insert(0,results[0])
                self.output1.insert(0,results[1])
                self.output2.insert(0,results[2])
                self.output3.insert(0,results[3])
                self.output4.insert(0,results[4])
                conn.close()


    def init_gui(self):
        """Builds GUI."""
        self.root.title('OPPLUS - Safiro V2 - Exportar Marcaciones ')
        self.root.option_add('*tearOff', 'FALSE')
 
        self.grid(column=0, row=0, sticky='nsew') # this starts the entire form

        # Input Boxes and Button           
        
        l1 = Label(self, text = "Ingrese Fecha de Inicio en formato AAAA-MM-DD:") 
        l2 = Label(self, text = "Ingrese Fecha de Corte en AAAA-MM-DD:") 
  
        # grid method to arrange labels in respective 
        # rows and columns as specified 
        l1.grid(row = 0, column = 0, sticky = W, pady = 2) 
        l2.grid(row = 1, column = 0, sticky = W, pady = 2) 
  
        # entry widgets, used to take entry from user 
        self.first_entry = tkinter.Entry(self, width=15) # first input box 
        self.last_entry = tkinter.Entry(self, width=15) # second input box 

        # this will arrange entry widgets 
        self.first_entry.grid(row = 0, column = 1, pady = 2) 
        self.last_entry.grid(row = 1, column = 1, pady = 2)       


        self.calc_button = ttk.Button(self, text='Obtener', command=self.calculate) # button
        self.calc_button.grid(column=2, row=1, columnspan=1, sticky='w', padx=14)

        # Output entries for answers

        self.output0 = tkinter.Entry(self, width=200, bd=0,)
        self.output0.grid(column=0, row=6, columnspan=5, padx=100, pady=0)
        self.output1 = tkinter.Entry(self, width=200, bd=0,)
        self.output1.grid(column=0, row=7, columnspan=5, padx=100, pady=0)
        self.output2 = tkinter.Entry(self, width=200, bd=0,)
        self.output2.grid(column=0, row=8, columnspan=5, padx=100, pady=0)
        self.output3 = tkinter.Entry(self, width=200, bd=0,)
        self.output3.grid(column=0, row=9, columnspan=5, padx=100, pady=0)
        self.output4 = tkinter.Entry(self, width=200, bd=0,)
        self.output4.grid(column=0, row=10, columnspan=5, padx=100, pady=0)

if __name__ == '__main__':
    root = tkinter.Tk()
    Adder(root)
    root.resizable(width=800, height=600) # locks window from being resized
    root.mainloop()
```
